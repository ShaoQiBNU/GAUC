```python
from sklearn.metrics import roc_auc_score
import pandas as pd
import numpy as np
import random
```

```python
a = [random.randint(0,9) for i in range(30)]
b = [random.randint(0,1) for i in range(30)]
c = [random.random() for i in range(30)]
d = [random.random()*10 for i in range(30)]

data = {"user_id": a, "label": b, "prob": c, "label2": d}
df = pd.DataFrame(data)
df
```

```python
def gauc(df, is_spearmanr = True):
  df_temp = df.groupby(['user_id'])
  roc = 0
  impression = 0
  for key, value in df_temp:
    if value.size <=1 or len(list(value.iloc[:, 1].unique())) <=1:
      continue
    else:
      if is_spearmanr:
        auc = roc_auc_score(value.iloc[:, 1], value.iloc[:, 2])
      else:
        _, auc = auc_custom(value.iloc[:, 1].tolist(), value.iloc[:, 2].tolist())
      
      roc += value.size * auc
      impression += value.size
  return roc / impression

def reversePairs(data):
    ans = 0
    prefix = []
    for n in data:
        left, right = 0, len(prefix)
        while left < right:
            mid = (left + right) // 2
            if n >= prefix[mid]:
                left = mid + 1
            else:
                right = mid
        ans += len(prefix) - left
        prefix[left:left] = [n]
    return ans

def auc_custom(labels, probs):
  
  n = len(labels)
  f = list(zip(probs, labels))

  rank_pos = [values2 for values1, values2 in sorted(f, key=lambda x: x[0], reverse=True)]
  pos_num = reversePairs(rank_pos)

  auc = pos_num / (n * (n - 1) / 2)

  return pos_num, auc
```

```python
gauc(df[["user_id", "label", "prob"]])
```
